```python
import functools
import jax
import os

from datetime import datetime
from jax import numpy as jp
import matplotlib.pyplot as plt

from IPython.display import HTML, clear_output

import brax

import flax
from brax import envs
from brax.io import model
from brax.io import json
from brax.io import html
```

```python
env_name = 'fetch'
env = envs.get_environment(env_name=env_name)
env_state = env.reset(rng=jax.random.PRNGKey(0))
# HTML(html.render(env.sys, [env_state.qp]))
```

```python
rollout = []
for i in range(30):  # You can adjust the number of simulation steps
    print(i)
    # Take a random action
    action = jax.random.uniform(jax.random.PRNGKey(0), shape=(env.action_size,))
    
    # Step the environment
    env_state = env.step(env_state, action)

    rollout.append(env_state.qp)
```

```python
HTML(html.render(env.sys, rollout))
```
